```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("Libraries loaded successfully!")
```

```python
# Load raw CSV
df_raw = pd.read_csv('..\data\dataset.csv', encoding='latin-1')

print(f"Shape: {df_raw.shape}")
print(f"\nFirst 10 rows:")
df_raw.head(10)
```

```python
# Check data types and missing values
print("Data Info:")
print(df_raw.info())
print("\n" + "="*50)
print("\nFirst few column names:")
print(df_raw.columns[:20].tolist())
```

```python
# Inspect first few rows to understand structure
print("Row 1 (Years):")
print(df_raw.iloc[1, :20].tolist())
print("\nRow 3 (Expenditure brackets):")
print(df_raw.iloc[3, :20].tolist())
print("\nRow 4 (Category - should be 'Makanan'):")
print(df_raw.iloc[4, 0])
print("\nRow 5 (First item - should be 'Padi-padian'):")
print(df_raw.iloc[5, 0])
```

```python
# Find which columns belong to which year
year_positions = {}
year_row = df_raw.iloc[1]  # Row with year info

current_year = None
for idx, val in enumerate(year_row):
    if pd.notna(val) and str(val).isdigit():
        current_year = int(val)
        if current_year not in year_positions:
            year_positions[current_year] = []
    if current_year is not None:
        year_positions[current_year].append(idx)

# Display year positions
for year in sorted(year_positions.keys()):
    print(f"{year}: {len(year_positions[year])} columns (cols {year_positions[year][0]}-{year_positions[year][-1]})")
```

```python
# Start from row 5 (actual data), skip row 4 ('Makanan' header)
# Column 0 is item names
df_clean = df_raw.iloc[5:].copy()
df_clean = df_clean.reset_index(drop=True)

# Keep only first column (item name) for now
df_clean = df_clean[[df_clean.columns[0]]]
df_clean.columns = ['item']

# Remove empty rows
df_clean = df_clean[df_clean['item'].notna() & (df_clean['item'] != '')]

# Remove summary rows
summary_keywords = ['Jumlah', 'Diolah dari', 'Sumber', 'Bukan Makanan']
df_clean = df_clean[~df_clean['item'].str.contains('|'.join(summary_keywords), case=False, na=False)]

print(f"Number of items extracted: {len(df_clean)}")
print("\nItem list:")
print(df_clean['item'].tolist())
```

```python
# Function to clean numeric values
def clean_numeric(val):
    if pd.isna(val) or val == '' or val == '�':
        return np.nan
    # Convert to string and clean
    val_str = str(val).strip()
    # Remove spaces
    val_str = val_str.replace(' ', '')
    # Handle comma as decimal separator
    if ',' in val_str:
        val_str = val_str.replace(',', '.')
    try:
        return float(val_str)
    except:
        return np.nan

# Extract data for each year
data_list = []

for year in sorted(year_positions.keys()):
    cols = year_positions[year]
    # Get expenditure brackets from row 3
    brackets = df_raw.iloc[3, cols].tolist()
    
    # Get data rows (starting from row 5)
    for idx, row_idx in enumerate(range(5, len(df_raw))):
        item = df_raw.iloc[row_idx, 0]
        
        # Skip if not valid item
        if pd.isna(item) or item == '' or any(kw in str(item) for kw in ['Jumlah', 'Diolah', 'Sumber', 'Bukan Makanan', 'Makanan']):
            continue
        
        # Extract values for this year
        values = df_raw.iloc[row_idx, cols].tolist()
        
        # Match brackets with values
        for bracket, value in zip(brackets, values):
            if pd.notna(bracket) and bracket != '':
                cleaned_value = clean_numeric(value)
                if pd.notna(cleaned_value):
                    data_list.append({
                        'year': year,
                        'item': item,
                        'expenditure_bracket': str(bracket),
                        'value': cleaned_value
                    })

# Create DataFrame
df_long = pd.DataFrame(data_list)

print(f"Total records extracted: {len(df_long)}")
print(f"\nYears: {sorted(df_long['year'].unique())}")
print(f"Number of years: {df_long['year'].nunique()}")
print(f"\nUnique items: {df_long['item'].nunique()}")
print(df_long['item'].unique()[:10])
print(f"\nUnique brackets: {df_long['expenditure_bracket'].nunique()}")
print(f"\nSample data:")
df_long.head(10)
```

```python
# Check data quality
print("Data Quality Check:")
print("="*50)
print(f"Total records: {len(df_long)}")
print(f"Missing values: {df_long['value'].isna().sum()}")
print(f"Duplicate records: {df_long.duplicated().sum()}")
print("\nValue statistics:")
print(df_long['value'].describe())
print("\nRecords per year:")
print(df_long.groupby('year').size())
```

```python
print("=== DATA QUALITY REPORT ===\n")
print(f"Shape: {df_long.shape}")
print(f"Missing values: {df_long.isnull().sum().sum()}")
print(f"Duplicate rows: {df_long.duplicated().sum()}")
print(f"\nValue statistics:")
print(df_long['value'].describe())
print(f"\nZero or negative values: {(df_long['value'] <= 0).sum()}")

# CHECKPOINT:
# - No missing values in year, item, bracket (value might have some)
# - Min value should be positive (>0)
# - Max value reasonable (<500,000)
```

```python
# Categorize food vs non-food
food_items = [
    'Padi-padian', 'Umbi-umbian', 'Ikan', 'Daging', 'Telur dan susu',
    'Sayur-sayuran', 'Kacang-kacangan', 'Buah-buahan', 'Minyak dan lemak',
    'Bahan minuman', 'Bumbu-bumbuan', 'Konsumsi lainnya', 'Makanan dan minuman jadi',
    'Tembakau dan sirih'
]

df_long['category'] = df_long['item'].apply(
    lambda x: 'Food' if any(food in x for food in food_items) else 'Non-Food'
)

print("=== ITEM CATEGORIZATION ===\n")
print(df_long.groupby('category')['item'].nunique())
print("\nFood items:")
print(df_long[df_long['category']=='Food']['item'].unique())
print("\nNon-Food items:")
print(df_long[df_long['category']=='Non-Food']['item'].unique())

# ✅ CHECKPOINT: Food items ~13-14, Non-Food items ~7-8
```

```python
# Standardize bracket names
def standardize_bracket(bracket):
    """Standardize expenditure bracket names"""
    bracket = str(bracket).strip()
    # Remove extra spaces
    bracket = ' '.join(bracket.split())
    return bracket

df_long['expenditure_bracket'] = df_long['expenditure_bracket'].apply(standardize_bracket)

print("=== EXPENDITURE BRACKETS ===\n")
print("Unique brackets:")
for bracket in sorted(df_long['expenditure_bracket'].unique()):
    count = (df_long['expenditure_bracket'] == bracket).sum()
    print(f"  {bracket}: {count:,} records")

# CHECKPOINT: Should have 8-9 distinct brackets with clean names
```

```python
# Create bracket order for sorting
bracket_order_map = {
    'Kurang dari 100 000': 1,
    '100 000- 149 999': 2,
    '150 000- 199 999': 3,
    '< 150 000': 3,
    '200 000- 299 999': 4,
    '300 000- 499 999': 5,
    '500 000- 749 999': 6,
    '750 000- 999 999': 7,
    '1 000 000- 1 499 999': 8,
    '1 000 000 dan lebih': 8,
    '> 1 500 000': 9
}

df_long['bracket_order'] = df_long['expenditure_bracket'].map(bracket_order_map)

# Fill missing with median
df_long['bracket_order'] = df_long['bracket_order'].fillna(df_long['bracket_order'].median())

print("=== BRACKET ORDERING ===\n")
print(df_long[['expenditure_bracket', 'bracket_order']].drop_duplicates().sort_values('bracket_order'))

# CHECKPOINT: All brackets should have order 1-9
```

```python
print("=== FINAL CLEANED DATA ===\n")
print(f"Shape: {df_long.shape}")
print(f"Columns: {df_long.columns.tolist()}")
print(f"\nData types:")
print(df_long.dtypes)
print(f"\nSample data:")
print(df_long.sample(10))

# CHECKPOINT: Should have columns: year, item, expenditure_bracket, value, category, bracket_order
```

```python
# Save to CSV
df_long.to_csv('../data/cleaned_data_long.csv', index=False)
print(" Data saved to: data/cleaned_data_long.csv")
print(f"   Records: {len(df_long):,}")
print(f"   Size: {df_long.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# CHECKPOINT: File should be created in data/ folder
```
